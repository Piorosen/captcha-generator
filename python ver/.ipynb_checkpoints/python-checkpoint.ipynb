```python
from captcha.image import ImageCaptcha
import os
import random as rn
```

```python
path = './Font'
fileList = os.listdir(path)

trainEpoch = 6
testEpoch = 1
savePath = './Save/'
trainPath = savePath + 'train/'
testPath = savePath + 'test/'
```

```python
for i in range(0, len(fileList)):
    fileList[i] = './Font/' +  fileList[i]
    
if not os.path.isdir(savePath):
    os.mkdir(savePath)
if not os.path.isdir(trainPath):
    os.mkdir(trainPath)
if not os.path.isdir(testPath):
    os.mkdir(testPath)
    
```

```python
image = ImageCaptcha(fonts=fileList)
```

```python
def RandomName(length):
    name = ""
    
    for i in range(0, length):
        n = rn.randint(0, 15)
        
        if (n > 9):
            name += chr(n + ord('a') - 10)
        else:
            name += chr(n + ord('0'))
            
    return name
    
```

```python
def HashName():
    name = ""
    name = RandomName(8) + '-' + RandomName(4) + '-' + RandomName(4) + '-' + RandomName(4) + '-' + RandomName(12)
    return name
```

```python
def Pick(origin, size, value = None):
    if value is None:
        value = []
    
    length = len(origin);
    
    if size == 0:
        Draw(value)

    for i in range(0, length):
        tmp = origin[i]
        
        value.append(tmp)
        del origin[i]
        Pick(origin, size - 1, value)
        origin.insert(i, tmp)
        value.pop()
```

```python
def ConvertNumToChar(num, value):
    for i in range(0, len(num) - len(value)):
        value = '0' + value
    
    charList = [['0', '영'], [1, '일'], [2, '이'], [3, '삼'], [4, '사'],
                [5, '오'], [6, '육'], [7, '칠'], [8, '팔'], [9, '구']]
    
    result = ""
    for i in range(0, len(num)):
        result += str(charList[int(num[i])][int(value[i])])
        
    return result
```

```python
def write(num, filePath, dec = 2):
    one = str(dec - 1) * len(num)
    one = one
    
    maxValue = int(one, dec)
    
    for i in range(0, maxValue):
        val = ConvertNumToChar(num, format(i, 'b'))
        name = num + '_' + HashName() + '.png'
        print(val)
        print(name)
        image.write(val, filePath + name);
    
```

```python
def ListToString(item):
    result = ""
    
    for i in item:
        result += chr(i + ord('0'))
        
    return result
```

```python
def Draw(item):
    num = ListToString(item)
    
    for i in range(0, trainEpoch):
        write(num, trainPath)
    for i in range(0, testEpoch):
        write(num, testPath)
    
```
